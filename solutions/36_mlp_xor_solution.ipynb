# 🔴 Solution: Train a 2-Layer MLP on XOR (Manual NumPy)

```python
import numpy as np
```

```python
# ✅ SOLUTION

def mlp_xor(X, Y, hidden_size=4, lr=0.1, epochs=10000) -> np.ndarray:
    input_size = X.shape[1]
    output_size = Y.shape[1]

    # He initialization
    W1 = np.random.randn(input_size, hidden_size) * np.sqrt(2.0 / input_size)
    b1 = np.zeros((1, hidden_size))
    W2 = np.random.randn(hidden_size, output_size) * np.sqrt(2.0 / hidden_size)
    b2 = np.zeros((1, output_size))

    for _ in range(epochs):
        # Forward
        Z1 = X.dot(W1) + b1
        A1 = np.maximum(0, Z1)  # ReLU
        Z2 = A1.dot(W2) + b2

        # MSE loss (not needed for updates, but useful for debugging)
        # loss = np.mean((Z2 - Y) ** 2)

        # Backward
        n = X.shape[0]
        dZ2 = (Z2 - Y) * 2 / n
        dW2 = A1.T.dot(dZ2)
        db2 = np.sum(dZ2, axis=0, keepdims=True)

        dA1 = dZ2.dot(W2.T)
        dZ1 = dA1 * (Z1 > 0).astype(float)  # ReLU derivative
        dW1 = X.T.dot(dZ1)
        db1 = np.sum(dZ1, axis=0, keepdims=True)

        # Update
        W1 -= lr * dW1
        b1 -= lr * db1
        W2 -= lr * dW2
        b2 -= lr * db2

    # Final predictions
    Z1 = X.dot(W1) + b1
    A1 = np.maximum(0, Z1)
    Z2 = A1.dot(W2) + b2
    return Z2
```

```python
# Verify
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0], [1], [1], [0]])
preds = mlp_xor(X, Y)
print("Predictions:", preds.ravel())
print("MSE:", np.mean((preds - Y) ** 2))
```
