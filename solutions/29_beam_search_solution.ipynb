# 🔴 Solution: Beam Search Decoding

```python
import torch
```

```python
# ✅ SOLUTION

def beam_search_decode(log_probs: torch.Tensor, beam_width: int) -> tuple[torch.Tensor, float]:
    beams = [([], 0.0)]
    seq_len, vocab_size = log_probs.shape

    for t in range(seq_len):
        candidates = []
        for sequence, score in beams:
            for token in range(vocab_size):
                candidates.append((sequence + [token], score + float(log_probs[t, token].item())))
        candidates.sort(key=lambda item: item[1], reverse=True)
        beams = candidates[:beam_width]

    best_sequence, best_score = beams[0]
    return torch.tensor(best_sequence, dtype=torch.long), best_score
```

```python
# Verify
log_probs = torch.log(torch.tensor([
    [0.6, 0.3, 0.1],
    [0.2, 0.5, 0.3],
], dtype=torch.float32))
print(beam_search_decode(log_probs, beam_width=2))
```
