# 🟢 Solution: K-Nearest Neighbors Classification

```python
import torch
```

```python
# ✅ SOLUTION

def knn_predict(x_train: torch.Tensor, y_train: torch.Tensor, x_query: torch.Tensor, k: int) -> torch.Tensor:
    distances = torch.cdist(x_query, x_train)
    nearest = distances.topk(k, largest=False, dim=-1).indices
    neighbor_labels = y_train[nearest]
    num_classes = int(y_train.max().item()) + 1
    preds = []
    for labels in neighbor_labels:
        preds.append(torch.bincount(labels, minlength=num_classes).argmax())
    return torch.stack(preds)
```

```python
# Verify
x_train = torch.tensor([[0.0, 0.0], [0.0, 1.0], [3.0, 3.0]])
y_train = torch.tensor([0, 0, 1])
x_query = torch.tensor([[0.1, 0.2], [2.8, 3.1]])
print(knn_predict(x_train, y_train, x_query, k=1))
```
