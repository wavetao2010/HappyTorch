# 🔴 Solution: PPO Clipped Policy Loss

```python
import torch
```

```python
# ✅ SOLUTION

def ppo_clipped_loss(old_log_probs, new_log_probs, advantages, clip_eps: float) -> torch.Tensor:
    ratio = torch.exp(new_log_probs - old_log_probs)
    unclipped = ratio * advantages
    clipped = torch.clamp(ratio, 1.0 - clip_eps, 1.0 + clip_eps) * advantages
    return -torch.minimum(unclipped, clipped).mean()
```

```python
# Verify
old_log_probs = torch.log(torch.tensor([0.2, 0.5, 0.3]))
new_log_probs = torch.log(torch.tensor([0.25, 0.45, 0.3]))
advantages = torch.tensor([1.0, -0.5, 2.0])
print(ppo_clipped_loss(old_log_probs, new_log_probs, advantages, clip_eps=0.2))
```
