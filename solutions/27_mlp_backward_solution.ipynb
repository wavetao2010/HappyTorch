# 🔴 Solution: Hand-Written Backward for a 2-Layer MLP

```python
import torch
```

```python
# ✅ SOLUTION

def mlp_backward(x, w1, b1, w2, b2, target) -> dict[str, torch.Tensor]:
    z1 = x @ w1.T + b1
    h = torch.relu(z1)
    y = h @ w2.T + b2

    grad_y = 2.0 * (y - target) / y.numel()
    grad_w2 = grad_y.T @ h
    grad_b2 = grad_y.sum(dim=0)

    grad_h = grad_y @ w2
    grad_z1 = grad_h * (z1 > 0).to(z1.dtype)
    grad_w1 = grad_z1.T @ x
    grad_b1 = grad_z1.sum(dim=0)

    return {
        "w1": grad_w1,
        "b1": grad_b1,
        "w2": grad_w2,
        "b2": grad_b2,
    }
```

```python
# Verify
x = torch.randn(2, 3)
w1 = torch.randn(4, 3)
b1 = torch.randn(4)
w2 = torch.randn(2, 4)
b2 = torch.randn(2)
target = torch.randn(2, 2)
grads = mlp_backward(x, w1, b1, w2, b2, target)
print(grads.keys())
```
