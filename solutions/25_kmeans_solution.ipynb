# 🟡 Solution: K-Means Clustering

```python
import torch
```

```python
# ✅ SOLUTION

def kmeans(x: torch.Tensor, num_clusters: int, num_iters: int = 10) -> tuple[torch.Tensor, torch.Tensor]:
    centroids = x[:num_clusters].clone()

    for _ in range(num_iters):
        distances = torch.cdist(x, centroids)
        assignments = distances.argmin(dim=1)

        new_centroids = centroids.clone()
        for cluster_idx in range(num_clusters):
            mask = assignments == cluster_idx
            if mask.any():
                new_centroids[cluster_idx] = x[mask].mean(dim=0)

        centroids = new_centroids

    distances = torch.cdist(x, centroids)
    assignments = distances.argmin(dim=1)
    return centroids, assignments
```

```python
# Verify
x = torch.tensor([
    [0.0, 0.0],
    [5.0, 5.0],
    [0.0, 1.0],
    [6.0, 5.0],
])
centroids, assignments = kmeans(x, num_clusters=2, num_iters=5)
print("Centroids:\n", centroids)
print("Assignments:", assignments)
```
