# 🟡 Solution: Temperature Sampling

```python
import torch
```

```python
# ✅ SOLUTION

def temperature_sample(logits: torch.Tensor, temperature: float) -> torch.Tensor:
    if temperature <= 0:
        raise ValueError("temperature must be positive")

    squeeze = logits.dim() == 1
    if squeeze:
        logits = logits.unsqueeze(0)

    probs = torch.softmax(logits / temperature, dim=-1)
    samples = torch.multinomial(probs, num_samples=1).squeeze(-1)
    return samples.squeeze(0) if squeeze else samples
```

```python
# Verify
torch.manual_seed(0)
logits = torch.tensor([[1.0, 2.0, 3.0], [3.0, 2.0, 1.0]])
print(temperature_sample(logits, temperature=0.7))
```
